```python
#converting words to numbers
input_seq = []
for sentence in dataset.split('\n'):
  tokenizer_sentence = tokenizer.texts_to_sequences([sentence])[0] #we added zero avoid a nested list

  for i in range(1, len(tokenizer_sentence)):
    input_seq.append(tokenizer_sentence[:i+1])
```

```python
input_seq #ideal input with sequences
```

```python
max_len = max([len(x) for x in input_seq])
```

```python
max_len
```

```python
#adding to padding to regularized the list size by adding zero padding
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
padded_input_sequences = pad_sequences(input_seq, maxlen = max_len, padding = 'pre') #padding successfully added #we used preffix padding cause we have to work on recent data and here our meaning full data alligns at the end
```

```python
#separting our input and output
x = padded_input_sequences[:,:-1]
y = padded_input_sequences[:,-1]
```

```python
x.shape
```

```python
#its better to make this a classification task as the regression would a continous ans and our dataset only has discrete token values
#we make this a multiclass classification by using onehotencoding
```

```python
len(tokenizer.word_index)
```

```python
from tensorflow.keras.utils import to_categorical
y = to_categorical(y,num_classes = 1714) #we put a extra number for onehot as it starts from zero not 1
```

```python
y.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
model = Sequential()
model.add(Embedding(1714, 100, input_length =  117))#outdenseis100, embedding layer is used to convert the sparse vector into dense vector
model.add(LSTM(150)) #number of nodes
model.add(Dense(1714, activation = 'softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
x.shape
```

```python
y.shape
```

```python
model.fit(x,y, epochs=100)
```
